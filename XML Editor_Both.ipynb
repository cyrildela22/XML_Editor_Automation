```python
#WORKING
import tkinter as tk
from tkinter import filedialog
import xml.etree.ElementTree as ET
import xml.dom.minidom as minidom
from tkinter import ttk


# Define dictionary for base fields
base_fields = {
    "NS1:AsOfDate": "value",
    "NS1:LastAnniversaryDate": "value",
    "NS1:PolNumber": "value",
    "NS1:Jurisdiction": "attribute",
    "NS1:EffDate": "value",
    "NS1:PaidToDate": "value",
    "NS1:PaymentMode": "attribute",
    "NS1:FaceAmt": "value",
    "NS1:IssueAge": "value",
    "NS1:IssueGender": "attribute",
    "NS1:TobaccoPremiumBasis": "attribute",
    "NS1:PermTableRating": "attribute",
    "NS1:UnderwritingClass": "attribute",
    "NS1:KeyValue": "value",
    "NS1:FirstName": "value",
    "NS1:LastName": "value",
    "NS1:Gender": "attribute",
    "NS1:BirthDate": "value",
    "NS1:AddressStateTC": "attribute"
}
# Define dictionary for additional fields
waiver_added_fields = {
    "NS1:CurrentAmt": "value",
    "NS1:AddedAge": "value",
}

# Dictionary to map the renamed keys to the original keys
renamed_field_values = {
    "As Of Date": "NS1:AsOfDate",
    "Last Anniversary Date": "NS1:LastAnniversaryDate",
    "Policy Number": "NS1:PolNumber",
    "Jurisdiction": "NS1:Jurisdiction",
    "Effective Date": "NS1:EffDate",
    "Paid To Date": "NS1:PaidToDate",
    "Payment Mode": "NS1:PaymentMode",
    "Face Amt": "NS1:FaceAmt",
    "Issue Age": "NS1:IssueAge",
    "Issue Gender": "NS1:IssueGender",
    "Tobacco Premium Basis": "NS1:TobaccoPremiumBasis",
    "Perm Table Rating": "NS1:PermTableRating",
    "Underwriting Class": "NS1:UnderwritingClass",
    "App Part 1 Date": "NS1:KeyValue",
    "First Name": "NS1:FirstName",
    "Last Name": "NS1:LastName",
    "Gender": "NS1:Gender",
    "BirthDate": "NS1:BirthDate",
    "AddressState TC": "NS1:AddressStateTC"
}
# Define dictionary to map additional fields
waiver_renamed_fields =   {
    "Current Amount": "NS1:CurrentAmt",
    "Added Age": "NS1:AddedAge"
}

#create a reverse mapping for renamed fields
reverse_renamed_field_values={v: k for k,v in renamed_field_values.items()}

#initialize dictionaries and variables
out_dict = {}
feature_dict = {}
global_path = []

user_input_value={}

#define a class for XML handling
class Xml_saver:
    
    #Define a method to browse and load an XML file
    def browse_xml_file(self):
        #         try:
        file_path = filedialog.askopenfilename(filetypes=[("XML Files", "*.xml")])
        global_path.append(file_path)
        if file_path:
            left = tk.Frame()
            left.pack(side='left', fill='x')
            right = tk.Frame()
            right.pack(side='right', fill='x') 
            drop_val = dropdown_var.get()
            if drop_val == "With WP":
        #                if not waiver_added_fields.keys()<=field_values.keys():
        #                    messagebox.showerror("Error","Selected template does not have Curr Amount and Added Age")
        #                return
                base_fields.update(waiver_added_fields)
                renamed_field_values.update(waiver_renamed_fields)
                self.field_values = base_fields
            else:
                self.field_values = base_fields
            out_dict = self.load_xml(file_path, self.field_values)

            for key in renamed_field_values.keys():
                # Use the renamed keys in the user interface
                #new_key = renamed_field_values.get(key, key)

                ori_key = renamed_field_values[key]
                label = tk.Label(left, text=key, font=('Arial', 10))
                #label = tk.Label(left, text=key, font=('Arial', 10))
                label.pack(side=tk.TOP)

                entry = tk.Entry(right, font=('Arial', 12))
                entry.pack(side=tk.TOP)
                entry.insert(0, out_dict[ori_key])
                feature_dict[ori_key] = entry
                user_input_value[ori_key] = out_dict[ori_key]
            print(user_input_value)
            out_dict[ori_key]
        #         except Exception as e:
        #             print("Error:", str(e))
        #             try:
        #                 tk.messagebox.showerror("Error", str(e))
        #             except Exception as messagebox_error:
        #                 print("Error displaying message:", str(messagebox_error))

#Define a method to load an XML file
    def load_xml(self, file_path, field_values):
        out_dict = {}
        try:
            #Parse XML file and extract data based on field_values
            all_keys = {}
            tree = ET.parse(file_path)
            print(tree)
            doc = minidom.parse(file_path)
            print(doc)
            
            
            # Check if the template is "With WP" and validate waiver_added_fields
            drop_val = dropdown_var.get()
            if drop_val == "With WP":
                missing_fields = []
                for key in waiver_added_fields.keys():
                    if not doc.getElementsByTagName(key):
                        missing_fields.append(key)
                if missing_fields:
                    error_msg = f"You indicated you are working on a policy that has WP, but you selected the xml template for a policy that does NOT have WP.  Please close out and reopen this application, and try again.."
                    tk.messagebox.showerror("Error", error_msg)
                    return out_dict
            elif drop_val == "No WP":
                invalid_fields = []
                for key in waiver_added_fields.keys():
                    if doc.getElementsByTagName(key):
                        invalid_fields.append(key)
                if invalid_fields:
                    error_msg = f"You indicated you are working on a policy that does NOT have WP, but you selected the xml template for a policy that has WP.  Please close out and reopen this application, and try again.."
                    tk.messagebox.showerror("Error", error_msg)
                    return out_dict
            return
            
            for key, val in field_values.items():
                # Find the elements with the field name
                #name = 'NS1:' + key
                name = key
                elements = doc.getElementsByTagName(name)
                for element in elements:
                    try:
                        if val == "value":

                            print(name, element.firstChild.data)
                            out_dict[name] = element.firstChild.data
                        else:
                            out_dict[name] = element.getAttribute('tc')
                    except:
                        print(f"Tag {name} not in XML file")


        except:
            print('Empty file')

        return out_dict 


#Define a method to save an XML file
    def save_xml(self):

        # Load XML file
        #print(global_path)
        xml_file = global_path[-1]
        doc = minidom.parse(xml_file)

        # Update the fields
        for field_name, field_entry in feature_dict.items():
            #print(field_name, field_entry.get())        
            elements = doc.getElementsByTagName(field_name)
            new_value = field_entry.get()
            for element in elements:
                dict_element = field_name[0:]
                if self.field_values[dict_element] == "value":
                    element.firstChild.nodeValue = new_value
                else:
                    element.setAttribute("tc", new_value)
            #user_input_value[ori_field_name] = new_value

        file_path = filedialog.asksaveasfilename(defaultextension=".xml", filetypes=[("XML Files", "*.xml")])
        if file_path:

            with open(file_path, 'w') as file:
                doc.writexml(file)

        window.quit

        return

#Create an instance of Xml_saver
saver = Xml_saver()

# Create the main window using tkinter
window = tk.Tk()
window.title("XML EDITOR")
window.geometry('350x550')


# Create a label
label = ttk.Label(text="Please select the template appropriate for your policy:")
label.pack( padx=10, pady=5)

#create a button to save changes
button_frame = tk.Frame(window)
button = tk.Button(button_frame, text="Save", command=saver.save_xml)
button.pack(side='bottom')

# Create a menu bar 
menu_bar = tk.Menu(window)
window.config(menu=menu_bar)

# Create a dropdown menu bar for template selection
dropdown_var = tk.StringVar()
#options = ["Waiver", "Without Waiver"]
options = ["With WP", "No WP"]
dropdown = ttk.Combobox(window, textvariable=dropdown_var, values=options, state="readonly",font=('Arial', 15))
dropdown.pack(padx=10, pady=10)
dropdown.current(1)

# Create the "File" menu and add commands
file_menu = tk.Menu(menu_bar, tearoff=0)
menu_bar.add_cascade(label="File", menu=file_menu)
file_menu.add_command(label="Open XML File", command=saver.browse_xml_file)
file_menu.add_command(label="Save XML File", command=saver.save_xml)
file_menu.add_separator()
file_menu.add_command(label="Exit", command=window.quit)

# Start the main event loop
window.mainloop()
#print(feature_dict)
```
